## Load Dataset

````python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("subhajournal/phishingemails")

print("Path to dataset files:", path)
````

````python
import os

os.listdir(path)
````

````python
import pandas as pd
import os

csv_file = os.path.join(path, "Phishing_Email.csv")

df = pd.read_csv(csv_file)

df.head()
````

## EDA

````python
df.dropna(subset=['Email Text'], inplace=True)
````

````python
df.isnull().sum()
````

````python
df['Email Type'].value_counts()
````

## Preprocessing

````python
df['Email Type'] = df['Email Type'].map(
    {    'Safe Email': 1,
    'Phishing Email': 0}
    )

df
````

## Tokenization and stopwords treatement

````python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
````

````python
df['Email Text'].isnull().sum()
````

````python
swords = stopwords.words('english')
````

````python
df['Email Text'] = df['Email Text'].apply(lambda x: x.split())
````

````python
cleaned = []

for words in df['Email Text']:

    new_words = []

    for word in words:
        if word.lower() not in swords:
            new_words.append(word)

    cleaned.append(new_words)

df['Email Text'] = cleaned
````

````python
df
````

````python
df['Email Text'] = df['Email Text'].apply(lambda x: ' '.join(x))
df
````

## Model Creation

### train test split

````python
from sklearn.model_selection import train_test_split
````

````python
X = df['Email Text'].values
y = df['Email Type'] .values
````

````python
X
````

````python
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size = 0.2, random_state=2)
````

### Vectorization

````python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer()

X_train = vect.fit_transform(X_train)
X_test = vect.transform(X_test)
````

### Model Fitting

````python
from sklearn.linear_model import LogisticRegression as lr
model = lr(max_iter=1000)
````

````python
model.fit(X_train, y_train)
````

## Model Evaluation

````python
from sklearn.metrics import accuracy_score as Acc
y_pred   = model.predict(X_test)
acc = Acc(y_test, y_pred)
print(acc)
````

````python
from sklearn.metrics import accuracy_score as Acc
y_pred   = model.predict(X_train)
acc = Acc(y_train, y_pred)
print(acc)
````

## Input Given

### Load Input and preprocessing

````python
ip = input("Enter the email: ")
ip = ip.split()
ip = [i for i in ip if i not in swords]
ip = ' '.join(ip)
````

### Predict output

````python
ip_vect = vect.transform([ip])
op = model.predict(ip_vect)
if op[0] == 0:
  print('Phishing Email')
else:
  print('Safe Email')
````
